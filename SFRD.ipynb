# BL_Sorini_HS_MD

```python
data2 = pd.read_csv("CSFRD_eta2.14_t2.39_ms2_mt2_mb2_tdyn1_Tfac1.0_Tmin4.0_Planck18_aL1_am1_aS1_ab1.dat", delim_whitespace=True, header=None)
data2.columns=['a','b']
data3 = pd.read_csv("sfr-output-010(g).txt", delim_whitespace=True, header=None)
data3.columns=['A','B']
data4 = pd.read_csv("CSFRD_HS03_Planck18_f1.txt", delim_whitespace=True, header=None)
data4.columns=['a','b']
data5 = pd.read_csv("CSFRD_eta1.9_t3.87_f1_ms2_mt2_mb2_tdyn1_Tfac1.0_Tmin4.5_Planck18.dat", delim_whitespace=True, header=None)
data5.columns=['a','b']

plt.xlim(20,1)
plt.ylim(0.005,7)

#Cosmological Paramters
omega_m=0.315
#omega_b=0.049
omega_l=0.684907
H0=0.0687087
H=67.4
h=H/100

#Density conversion
Rho_crit=(2.7754*10**11)*(h**2)
Rho_m=(omega_m*Rho_crit)/10**9
C=Rho_m*(data3['B']) 

#Conversion of Time to Redshift
time=(data3['A'])
def a(time) :
    return (omega_m/omega_l)**(1/3)*(np.sinh((3/2)*(math.sqrt(omega_l)*H0*time))**(2/3))
#Redshift
z=(a(time)**-1)-1

#Equation 15 from Madau Dickinson paper (M&D Fit)
def Psi(C):
    return (0.015*(1+z)**2.7)/(1+((1+z)/2.9)**5.6)
Psi(C)

#Plotting of graph
plt.plot(z+1,np.array(Psi(C)),'--',color='grey')
plt.plot(np.array(data2['a']+1),np.array(data2['b']),color='green')
plt.plot(np.array(data4['a']+1),np.array(data4['b']),'--',color='blue')
plt.plot(z+1,np.array(C),'--',color='r')
plt.plot(np.array(data5['a']+1),np.array(data5['b']),color='black')
plt.xlabel('Redshift [Z+1]')
plt.ylabel(r'SFRD [$M_{\odot}yr^{-1}Mpc^{-3}]$')
plt.legend(["Madau & Dickinson (2014) Fit ","Sorini - Fiducial Parameters","Hernquist & Springel (2003) ","Bousso & Leichenauer","Sorini - BestFit Parameters"])
plt.title(r'$\alpha m=1 ; \alpha \lambda=1 ; \alpha b=1; \alpha s=1 $')
#plt.grid()
#plt.savefig('Comparison BL_Sorini Standard',dpi=400)

plt.yscale('log')
plt.xscale('log')
plt.savefig('BL_Sorini_HS_MD',dpi=400)
```

# SFRD Comparison-MD14 / ratio comparison

```python
#Sorini-Fiducial Parameters
C1=Rho_m*(data2['b']) 
def Psi(C1):
    return (0.015*(1+(data2['a']))**2.7)/(1+((1+(data2['a']))/2.9)**5.6)
D=(data2['b']/Psi(C1))
E=(data2['a'])
plt.plot(E+1,D,'--',color='green',label='Sorini & Peacock -Fiducial Parameters')

#Hernquist & Springel (2003)
C2=Rho_m*(data4['b']) 
def Psi(C2):
    return (0.015*(1+(data4['a']))**2.7)/(1+((1+(data4['a']))/2.9)**5.6)
F=(data4['b']/Psi(C2))
G=(data4['a'])
plt.plot(G+1,F,'--',color='blue',label='Hernquist & Springel (2003)')

#Sorini-Best Fit Parameters
C3=Rho_m*(data5['b']) 
def Psi(C3):
    return (0.015*(1+(data5['a']))**2.7)/(1+((1+(data5['a']))/2.9)**5.6)
H=(data5['b']/Psi(C3))
I=(data5['a'])
plt.plot(I+1,H,'--',color='black',label='Sorini & Peacock -Best Fit Parameters')

#Bousso & Leichenauer
def Psi(C):
    return (0.015*(1+z)**2.7)/(1+((1+z)/2.9)**5.6)
Psi(C)
plt.plot(z+1,np.array(C)/np.array(Psi(C)),'--',color='red',label='Bousso & Leichenauer')

plt.xlabel('Redshift [Z+1]')
plt.ylabel(r'$\rho_*/\rho_*^{MD14}$')
plt.legend()
plt.title(r'$\alpha m=1 ; \alpha \lambda=1 ; \alpha b=1; \alpha s=1 $')
plt.xlim(10,1)
plt.ylim(0.005,40)
plt.yscale('log')
plt.xscale('log')
plt.savefig('SFRD Comparison-MD14',dpi=400)
```
